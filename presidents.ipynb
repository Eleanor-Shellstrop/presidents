```python
import pandas as pd
```

```python
df = pd.read_csv("datasets/lifespan.csv")
```

```python
df.head(10)
```

```python
df['dob_year'] = df['dob'].str[-4:]
df.head()
```

```python
df['dod_year'] = df['dod'].str[-4:]
df.head()
```

```python
df[['birth_city', 'birth_state']] = df['birth_city'].str.split(',', expand=True)
df.head()
```

```python
df[['death_city', 'death_state']] = df['death_city'].str.split(',', expand=True)
df.head()
```

```python
cols = ['birth_state', 'death_state']

for col in cols:
    r = df[col].str.replace
    df[col] = r('Cal.', 'California', regex=True)
    df[col] = r('Conn.', 'Connecticut', regex=True)
    df[col] = r('Ga.', 'Georgia', regex=True)
    df[col] = r('MA', 'Massachusetts', regex=True)
    df[col] = r('Mass.', 'Massachusetts', regex=True)
    df[col] = r('N.C.', 'North Carolina', regex=True)
    df[col] = r('New Hamp.', 'New Hampshire', regex=True)
    df[col] = r('N.H.', 'New Hampshire', regex=True)
    df[col] = r('No/So Carolina', 'South Carolina', regex=True)
    df[col] = r('Pa.', 'Pennsylvania', regex=True)
    df[col] = r('Tenn.', 'Tennessee', regex=True)
    df[col] = r('Va.', 'Virginia', regex=True)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df)
```

```python
df['middle_name'] = df['middle_name'].fillna('')
df.head(10)
```

```python
df['full_name'] = df['first_name'] + " " + df['middle_name'] + " " + df['last_name']
df.head()
```

```python
df['full_name'] = df['full_name'].str.replace("  ", " ")
```

```python
df.to_csv("datasets/updated_lifespan.csv")
```
